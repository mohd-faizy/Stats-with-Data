```python
# Importing the necessary module
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import statistics as stats
%matplotlib inline 
```

```python
# Loading the tips dataset
df_tips = sns.load_dataset('tips')
df_tips
```

```python
# Mean, Median & mode
print(np.mean(df_tips['total_bill']))
print(np.median(df_tips['total_bill']))
print(stats.mode(df_tips['total_bill']))
```

```python
# summary statistics
df_tips.describe()
```

```python
# Box Plot
sns.boxplot(x = df_tips['total_bill'])
plt.show()
```

```python
sns.histplot(x=df_tips['total_bill'], kde=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

fig, axs = plt.subplots(ncols=4, figsize=(18, 6))

sns.boxplot(data=iris, x='species', y='sepal_length', ax=axs[0])
sns.boxplot(data=iris, x='species', y='sepal_width', ax=axs[1])
sns.boxplot(data=iris, x='species', y='petal_length', ax=axs[2])
sns.boxplot(data=iris, x='species', y='petal_width', ax=axs[3])

plt.show()
```
