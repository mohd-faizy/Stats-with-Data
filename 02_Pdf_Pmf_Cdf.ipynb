#  **PMF** vs **PDF** vs **CDF** 

Here is the difference between PMF, PDF and CDF:

**PMF** stands for Probability Mass Function. It represents the probability of a discrete random variable taking on a particular value. PMF indicates the probability that a discrete variable will equal a specific possible outcome.

**PDF** stands for Probability Density Function. It represents the probability of a continuous random variable falling within a certain range of values. PDF indicates the likelihood that a continuous variable will take on values in a particular interval.

**CDF** stands for Cumulative Distribution Function. It represents the probability that a random variable will be less than or equal to a specific value. CDF shows the cumulative probability over the range of values.

Some key differences:

-  PMF is used for discrete random variables while PDF is used for continuous random variables.
-  PMF returns a probability mass at each point value while PDF returns the probability density over a range of values.
-  The sum of probabilities from a PMF is 1 while the integral of probabilities from a PDF is 1.
-  A CDF shows the total probability up to a value whereas PMF and PDF show the probability at or within a value.
-  PMF is defined as P(X=x) while PDF is defined as f(x) and CDF is defined as F(x).

`PMF`, `PDF`, and `CDF` are all ways to describe the distribution of a `random variable

| **PMF** | **PDF** | **CDF** |
| :---: | :---: | :---: |
| Probability mass function | Probability density function | Cumulative distribution function |
| Discrete random variable | Continuous random variable | Discrete or continuous random variable |
| Probability assigned to each possible value | Probability assigned to an interval of values | Probability assigned to a range of values |
| $P(X = x)$ | $f(x)$ | $F(x) = P(X \leq x)$ |
| $0 \leq P(X = x) \leq 1$ | $f(x) \geq 0$ | $0 \leq F(x) \leq 1$ |
| $\sum_{x}P(X = x) = 1$ | $\int_{-\infty}^{\infty}f(x)dx = 1$ | $\lim_{x \rightarrow -\infty}F(x) = 0$, $\lim_{x \rightarrow \infty}F(x) = 1$ |


## **PDF**



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the mean and standard deviation of the normal distribution
mu = 0
sigma = 1

# Create an array of x-values to plot the PDF
x = np.linspace(mu - 3*sigma,
                mu + 3*sigma,
                100)

# Calculate the PDF using the norm() function from the scipy.stats module
pdf = norm.pdf(x, mu, sigma)

# Plot the PDF
plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('Probability density')
plt.title(f'Normal distribution with mean = {mu} and standard deviation = {sigma}')
plt.show()
```

```python
import math

def normal_pdf(x, mu=0, sigma=1):
    """
    Returns the value of the normal distribution's PDF for a given x, mean, and standard deviation.
    """
    return (1 / math.sqrt(2 * math.pi * sigma**2)) * math.exp(- (x - mu)**2 / (2 * sigma**2))

# Example usage
x_values = [i / 10.0 for i in range(-50, 51)]  # Generate a list of x values from -5 to 5
y_values = [normal_pdf(x) for x in x_values]  # Generate a list of y values using the PDF formula

# Plot the PDF using matplotlib
import matplotlib.pyplot as plt

plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Normal Distribution PDF')
plt.show()
```

## **CDF**



```python
import math

def normal_cdf(x, mu=0, sigma=1):
    """
    Returns the value of the normal distribution's CDF for a given x, mean, and standard deviation.
    """
    return (1/2) * (1 + math.erf((x - mu) / (sigma * math.sqrt(2))))

# Example usage
x_values = [i / 10.0 for i in range(-50, 51)]  # Generate a list of x values from -5 to 5
y_values = [normal_cdf(x) for x in x_values]  # Generate a list of y values using the CDF formula

# Plot the CDF using matplotlib
import matplotlib.pyplot as plt

plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('CDF')
plt.title('Normal Distribution CDF')
plt.show()
```

## **PMF**

```python
import matplotlib.pyplot as plt
import math
import numpy as np


def normal_pmf(x_values, mu=0, sigma=1, num_bins=101):
    """
    Returns the PMF of a discretized Normal distribution for given x values, mean, standard deviation, and number of bins.
    """
    # Divide the range of x_values into num_bins equally spaced bins
    bins = np.linspace(min(x_values), max(x_values), num_bins + 1)

    # Calculate the probability mass of each bin using the Normal PDF formula and binning
    bin_probs = []
    for i in range(num_bins):
        bin_start = bins[i]
        bin_end = bins[i+1]
        bin_width = bin_end - bin_start
        bin_center = (bin_start + bin_end) / 2
        bin_prob = normal_pdf(bin_center, mu, sigma) * bin_width
        bin_probs.append(bin_prob)

    # Normalize the bin probabilities so that they sum up to 1
    bin_probs = np.array(bin_probs)
    bin_probs /= np.sum(bin_probs)

    return bin_probs


# Example usage
num_bins = 10  # Set the number of bins to use for the PMF calculation
# Generate a list of x values from -50 to 50 with num_bins equally spaced values
x_values = np.linspace(-5, 5, num_bins)
# Generate the PMF using binning
pmf_values = normal_pmf(x_values, mu=0, sigma=1, num_bins=num_bins)

# Plot the PMF using matplotlib

plt.bar(x_values, pmf_values, width=1)
plt.xlabel('x')
plt.ylabel('PMF')
plt.title('Discretized Normal Distribution PMF')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Define the probability mass function (PMF) for a six-sided die
pmf = np.ones(6) / 6
print(pmf)

# This code generates an array of 100 evenly spaced numbers between -3 and 3 using the  `linspace`  function in NumPy.
x = np.linspace(-3, 3, 100)
print(x[:10])
sns.histplot(x)
plt.show()

# Define the probability density function (PDF) for a normal distribution
pdf = norm.pdf(x)

# Define the cumulative distribution function (CDF) for a normal distribution
cdf = norm.cdf(x)

# Create a bar plot of the PMF
plt.subplot(3, 1, 1)
plt.bar(np.arange(1, 7), pmf)
plt.title("PMF for a six-sided die")

# Create a line plot of the PDF
plt.subplot(3, 1, 2)
plt.plot(x, pdf)
plt.title("PDF for a standard normal distribution")

# Create a line plot of the CDF
plt.subplot(3, 1, 3)
plt.plot(x, cdf)
plt.title("CDF for a standard normal distribution")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

# Define the probability mass function (PMF) for a six-sided die
pmf = np.ones(6) / 6

# Define the probability density function (PDF) for a normal distribution
pdf = lambda x: np.exp(-x**2 / 2) / np.sqrt(2*np.pi)

# Define the cumulative distribution function (CDF) for a normal distribution
cdf = lambda x: (1 + erf(x / np.sqrt(2))) / 2

# Create a bar plot of the PMF
plt.subplot(3, 1, 1)
plt.bar(np.arange(1, 7), pmf)
plt.title("PMF for a six-sided die")

# Create a line plot of the PDF
x = np.linspace(-5, 5, 100)
plt.subplot(3, 1, 2)
plt.plot(x, pdf(x))
plt.title("PDF for a standard normal distribution")

# Create a line plot of the CDF
plt.subplot(3, 1, 3)
plt.plot(x, cdf(x))
plt.title("CDF for a standard normal distribution")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a random dataset
data = np.random.normal(0, 1, size=10000)

# Calculate the PMF
pmf, edges = np.histogram(data, bins=20, density=True)
pmf_df = pd.DataFrame({'value': pmf, 'left': edges[:-1], 'right': edges[1:]})

# Calculate the PDF
pdf_x = np.linspace(data.min(), data.max(), 100)
pdf_y = np.exp(-(pdf_x - 0)**2 / 2) / np.sqrt(2*np.pi)
pdf_df = pd.DataFrame({'value': pdf_y, 'x': pdf_x})

# Calculate the CDF
cdf_x = np.sort(data)
cdf_y = np.arange(1, len(cdf_x) + 1) / len(cdf_x)
cdf_df = pd.DataFrame({'value': cdf_y, 'x': cdf_x})

# Plot the PMF
plt.subplot(3, 1, 1)
plt.bar(pmf_df['left'], pmf_df['value'], width=(pmf_df['right'] - pmf_df['left']))
plt.title("PMF for a normal distribution")

# Plot the PDF
plt.subplot(3, 1, 2)
plt.plot(pdf_df['x'], pdf_df['value'])
plt.title("PDF for a normal distribution")

# Plot the CDF
plt.subplot(3, 1, 3)
plt.plot(cdf_df['x'], cdf_df['value'])
plt.title("CDF for a normal distribution")

plt.tight_layout()
plt.show()
```

## **for IRIS Dataset**

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Select only the setosa species
setosa_petal_length = iris.loc[iris['species'] == 'setosa', 'petal_length']

# Calculate the PMF of petal length
pmf, edges = np.histogram(setosa_petal_length, bins=10, density=True)
pmf_df = pd.DataFrame({'value': pmf, 'left': edges[:-1], 'right': edges[1:]})

# Plot the PMF
plt.bar(pmf_df['left'], pmf_df['value'], width=(pmf_df['right'] - pmf_df['left']))
plt.title("PMF of Petal Length for Setosa Iris")
plt.xlabel("Petal Length")
plt.ylabel("Probability Density")
plt.show()
```

```python
from scipy.stats import norm

# Fit a normal distribution to the petal length data
mu, std = norm.fit(setosa_petal_length)

# Create a PDF of petal length using the fitted normal distribution
pdf_x = np.linspace(setosa_petal_length.min(), setosa_petal_length.max(), 100)
pdf_y = norm.pdf(pdf_x, mu, std)
pdf_df = pd.DataFrame({'value': pdf_y, 'x': pdf_x})

# Plot the PDF
plt.plot(pdf_df['x'], pdf_df['value'])
plt.title("PDF of Petal Length for Setosa Iris")
plt.xlabel("Petal Length")
plt.ylabel("Probability Density")
plt.show()
```

```python
# Create a CDF of petal length
cdf_x = np.sort(setosa_petal_length)
cdf_y = np.arange(1, len(cdf_x) + 1) / len(cdf_x)
cdf_df = pd.DataFrame({'value': cdf_y, 'x': cdf_x})

# Plot the CDF
plt.plot(cdf_df['x'], cdf_df['value'])
plt.title("CDF of Petal Length for Setosa Iris")
plt.xlabel("Petal Length")
plt.ylabel("Cumulative Probability")
plt.show()
```
