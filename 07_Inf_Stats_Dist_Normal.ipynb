```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data from a normal distribution with mean 0 and standard deviation 1
data = np.random.normal(0, 1, 1000)

# Calculate the mean and standard deviation of the data
mean = np.mean(data)
std = np.std(data)

# Create a histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.5, label='Data')

# Create a normal distribution with the same mean and standard deviation as the data
x = np.linspace(-4, 4, 100)
y = 1/(std * np.sqrt(2 * np.pi)) * np.exp(-(x - mean)**2 / (2 * std**2))
plt.plot(x, y, 'r-', label='Normal Distribution')

# Add axis labels and a legend
plt.xlabel('Data Values')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

## Normal Distribution of Total Bill in Tips Dataset

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load the datasets
tips = sns.load_dataset("tips")

# Tips normal distribution plot
plt.figure(figsize=(10, 5))
sns.histplot(tips['total_bill'], kde=True, label='Total Bill')
plt.title("Normal Distribution of Total Bill in Tips Dataset")
plt.legend()
plt.show()
```

## Normal Distribution of Sepal Length in Iris Dataset

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

iris = sns.load_dataset("iris")

# Iris normal distribution plot
plt.figure(figsize=(10, 5))
sns.histplot(x=iris['sepal_length'], kde=True, label='Sepal Length')
plt.title("Normal Distribution of Sepal Length in Iris Dataset")
plt.legend()
plt.show()
```
